```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/TomerStupko/data-analysis-project/refs/heads/main/Student%20Insomnia%20and%20Educational%20Outcomes%20Dataset_version-2.csv")
df
```

3. `?האם חוסר בשעות שינה מגביר את הלחץ בלימודים`

```python
df.columns
```

```python
df = df.rename(columns={'1. What is your year of study?':'year of study','2. What is your gender?':'gender','3. How often do you have difficulty falling asleep at night? ':'difficulty falling asleep','4. On average, how many hours of sleep do you get on a typical day?':'sleep hours','5. How often do you wake up during the night and have trouble falling back asleep?':'wake up at night','6. How would you rate the overall quality of your sleep?':'sleep quality','7. How often do you experience difficulty concentrating during lectures or studying due to lack of sleep?':'difficulty concentrating','8. How often do you feel fatigued during the day, affecting your ability to study or attend classes?':'fatigued during the day','9. How often do you miss or skip classes due to sleep-related issues (e.g., insomnia, feeling tired)?':'skip class','10. How would you describe the impact of insufficient sleep on your ability to complete assignments and meet deadlines?':'impact on ability to complete assignments','11. How often do you use electronic devices (e.g., phone, computer) before going to sleep?':'screen time before bed','12. How often do you consume caffeine (coffee, energy drinks) to stay awake or alert?':'caffeine consumption','13. How often do you engage in physical activity or exercise?':'physical activity','14. How would you describe your stress levels related to academic workload?':'academic stress level','15. How would you rate your overall academic performance (GPA or grades) in the past semester?':'academic performance'})
df.columns
```

```python
df.head()
```

```python
df['year of study_encoded'] = df['year of study'].map({
    'First year': 1,
    'Second year': 2,
    'Third year': 3,
    'Graduate student': 4,
})
df['difficulty falling asleep_encoded'] = df['difficulty falling asleep'].map({
    'Never': 0,
    'Rarely (1-2 times a week)': 1,
    'Sometimes (3-4 times a week)': 2,
    'Often (5-6 times a week)': 3,
    'Every night': 4
})

df['sleep hours_encoded'] = df['sleep hours'].map({
    'Less than 4 hours': 0,
    '4-5 hours': 1,
    '6-7 hours': 2,
    '7-8 hours': 3,
    'More than 8 hours': 4
})

df['wake up at night_encoded'] = df['wake up at night'].map({
    'Never': 0,
    'Rarely (1-2 times a week)': 1,
    'Sometimes (3-4 times a week)': 2,
    'Often (5-6 times a week)': 3,
    'Every night': 4
})

df['sleep quality_encoded'] = df['sleep quality'].map({
    'Very poor': 0,
    'Poor': 1,
    'Average': 2,
    'Good': 3,
    'Very good': 4
})

df['difficulty concentrating_encoded'] = df['difficulty concentrating'].map({
    'Never': 0,
    'Rarely': 1,
    'Sometimes': 2,
    'Often': 3,
    'Always': 4
})

df['fatigued during the day_encoded'] = df['fatigued during the day'].map({
    'Never': 0,
    'Rarely': 1,
    'Sometimes': 2,
    'Often': 3,
    'Always': 4
})

df['skip class_encoded'] = df['skip class'].map({
    'Never': 0,
    'Rarely (1-2 times a month)': 1,
    'Sometimes (1-2 times a week)': 2,
    'Often (3-4 times a week)': 3,
    'Always': 4
})

df['impact on ability to complete assignments_encoded'] = df['impact on ability to complete assignments'].map({
    'No impact': 0,
    'Minor impact': 1,
    'Moderate impact': 2,
    'Major impact': 3,
    'Severe impact': 4
})

df['screen time before bed_encoded'] = df['screen time before bed'].map({
    'Never': 0,
    'Rarely (1-2 times a week)': 1,
    'Sometimes (3-4 times a week)': 2,
    'Often (5-6 times a week)': 3,
    'Every night': 4
})

df['caffeine consumption_encoded'] = df['caffeine consumption'].map({
    'Never': 0,
    'Rarely (1-2 times a week)': 1,
    'Sometimes (3-4 times a week)': 2,
    'Often (5-6 times a week)': 3,
    'Every day': 4
})

df['physical activity_encoded'] = df['physical activity'].map({
    'Never': 0,
    'Rarely (1-2 times a week)': 1,
    'Sometimes (3-4 times a week)': 2,
    'Often (5-6 times a week)': 3,
    'Every day': 4
})

df['academic stress level_encoded'] = df['academic stress level'].map({
    'No stress': 0,
    'Low stress': 1,
    'High stress': 2,
    'Extremely high stress': 3
})

df['academic performance_encoded'] = df['academic performance'].map({
    'Poor': 0,
    'Below Average': 1,
    'Average': 2,
    'Good': 3,
    'Excellent': 4
})

df.head()
```

```python
plt.figure(figsize=(8,4))
correlation_matrix = df.corr(numeric_only = True,method = 'pearson').round(2)
sns.heatmap(data=correlation_matrix,cmap='coolwarm', annot=True)
```

### We created a graph that shows the distribution of different levels of fatigue across various caffeine consumption levels.

```python
plt.figure(figsize=(6, 3))
order = ['Never','Rarely (1-2 times a week)','Sometimes (3-4 times a week)','Often (5-6 times a week)','Every day']
colors = sns.color_palette("Oranges", n_colors=len(order))
color_dict = dict(zip(order, colors))
sns.barplot(data = df,x = 'caffeine consumption',hue = 'caffeine consumption',order=order, y = 'fatigued during the day_encoded',errorbar = None,palette=color_dict,legend=False)
plt.xticks(rotation = 25)
plt.show()
```

#### As the graph shows, those who reported they consume more caffeine, also reported that they feel more fatigue.

#### We wanted to see if the conclusion still holds when considering only those whose sleep quality was good.

```python
fig, axes = plt.subplots(figsize=(15, 3), ncols=2)
order = ['Never','Rarely (1-2 times a week)','Sometimes (3-4 times a week)','Often (5-6 times a week)','Every day']
colors = sns.color_palette("Oranges", n_colors=len(order))
color_dict = dict(zip(order, colors))
sns.set(style="whitegrid")
sns.barplot(data = df,x = 'caffeine consumption',order=order, y = 'fatigued during the day_encoded', ax = axes[0],hue = 'caffeine consumption',errorbar = None,palette=color_dict)
sns.barplot(data = df[df['sleep quality_encoded']>2],x = 'caffeine consumption',order=order, y = 'fatigued during the day_encoded',hue = 'caffeine consumption', ax = axes[1],errorbar = None,palette=color_dict)
axes[0].set_title("Caffeine Consumption vs. fatigue (All Participants)", fontsize=14)
axes[0].set_xlabel("Caffeine Consumption (Encoded)", fontsize=12)
axes[1].set_title("Caffeine consumption vs. Fatigue (Sleep Quality good)", fontsize=14)
axes[1].set_xlabel("Caffeine Consumption (Encoded)", fontsize=12)
axes[0].tick_params(axis='x', labelrotation=45)
axes[1].tick_params(axis='x', labelrotation=45)
plt.show()
```

```python
sns.set(style="whitegrid") 
plt.figure(figsize=(6, 4))
sns.lineplot(data=df,x='caffeine consumption_encoded',y='academic performance_encoded',errorbar=None,marker='o',linewidth=2.5,color="#5a7fa3")
plt.title("Caffeine Consumption vs. Academic Performance", fontsize=16, fontweight='bold')
plt.xlabel("Caffeine Consumption (Encoded)", fontsize=12)
plt.ylabel("Academic Performance (Encoded)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid") 
fig, axes = plt.subplots(figsize=(12, 4), ncols=2)
sns.lineplot(data=df, x='caffeine consumption_encoded',y='academic performance_encoded',errorbar=None,marker='o',linewidth=2.5,ax=axes[0])
axes[0].set_title("All Participants", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Caffeine Consumption (Encoded)", fontsize=11)
axes[0].set_ylabel("Academic Performance (Encoded)", fontsize=11)
sns.lineplot(data=df[df['sleep quality_encoded'] > 2],x='caffeine consumption_encoded',y='academic performance_encoded', errorbar=None,marker='o',linewidth=2.5,ax=axes[1])
axes[1].set_title("Good Sleep Quality Only", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Caffeine Consumption (Encoded)", fontsize=11)
axes[1].set_ylabel("Academic Performance (Encoded)", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
order = ['Never','Rarely (1-2 times a week)','Sometimes (3-4 times a week)','Often (5-6 times a week)','Every day']
sns.boxplot(data = df, x = 'physical activity', y = 'academic stress level_encoded',order = order,medianprops=dict(color='black', linewidth=1.5))
plt.xticks(rotation = 25)
plt.show()
```

### As it shown, people who reported that they do more physical activity, also reported higher stress levels.

### But is there a connection between the academic stress and difficulty concentrating?

```python
sns.regplot(data = df, x = 'difficulty concentrating_encoded', y = 'academic stress level_encoded')
```

#### We can see that there is a modest connection between difficulty concentrating and academic stress (0.32).

```python
sns.set(style="whitegrid")
order = ['Never','Rarely (1-2 times a week)','Sometimes (3-4 times a week)','Often (5-6 times a week)','Every day']
colors = sns.color_palette("Blues", n_colors=len(order))
color_dict = dict(zip(order, colors))
g = sns.FacetGrid(df, col='difficulty concentrating_encoded', col_wrap=5, height=4, sharey=True)
g.map_dataframe(sns.barplot,x='physical activity',y='academic stress level_encoded',hue='physical activity',order=order,errorbar=None,palette=color_dict,legend=False)
for ax in g.axes.flat:
    ax.tick_params(axis='x', rotation=45)
    ax.set_xlabel('physical activity')
    ax.set_ylabel('Academic Stress')
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle("Academic Stress by Physical Activity and Concentration Difficulty", fontsize=16, fontweight='bold')
plt.show()
```

#### As shown in the graph, the more time people spent on physical activity, the higher their academic stress became across all levels of concentration difficulty. In other words, regardless of how much difficulty they had concentrating, there is still a strong relationship between physical activity and academic stress.
